```python
import numpy as np
```

```python
# one unit matrix, all zeros 13x15
unit_mat_13x25 = np.zeros((3,13,25), dtype=int)
```

```python
# create a single row with 13 13x15 matrices
unit_row_13x750_list = []

# list of 10 13x15 matrices
for i in range(30):
  unit_row_13x750_list.append(unit_mat_13x25 + i)
```

```python
# create a single row for one matrix
unit_row_13x750 = np.concatenate((tuple(unit_row_13x750_list)), axis=2)
```

```python
unit_row_13x750.shape
```

```python
# create multiple rows to create one base matrix
# Should be 10 instances of rows
# need to increment the values of each row by 30
# [0, 30,..., ]

unit_matrix_130x750_list = []
for i in np.arange(0,300,30):
  unit_matrix_130x750_list.append(unit_row_13x750 + i)
```

```python
len(unit_matrix_130x750_list)
```

```python
unit_matrix_130x750 = np.concatenate(tuple(unit_matrix_130x750_list), axis=1)
```

```python
unit_matrix_130x750.shape
```

```python
# sanity check
unit_matrix_130x750[1,12,:] # then when you do unit_matrix_130x750[1,13,:] you'll get everything incremented by 30
```

```python
# need 4 base matrices to create one whole mask
# want 4 x 3 x 130 x 750
```

```python
full_matrix_4x3x130x750_list = []
for i in np.arange(0, 1200, 300):
  full_matrix_4x3x130x750_list.append(unit_matrix_130x750 + i)
```

```python
len(full_matrix_4x3x130x750_list)
```

```python
full_matrix_4x3x130x750_list[1]
```

```python
full_matrix_3x4x130x750 = np.zeros((4,3,130,750))
```

```python
for i in range(len(full_matrix_4x3x130x750_list)):
  full_matrix_3x4x130x750[i] = full_matrix_4x3x130x750_list[i]
```
